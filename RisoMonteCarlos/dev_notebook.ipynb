```python
import numpy as np
import matplotlib as matplt
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

paths=[]
ims=[]
for i in range(1,9):
    path='riso'+f"{i}"+'.png'
    paths.append(path)
    #ims.append(image.imread(path))
    #image = plt.imread(path)[116:116+30, 236:236+30]
    ims.append(plt.imread(path)) #[6:6+30, 126:126+30])

def plot_images(x, y, ims, ax=None):
    ax = ax or plt.gca()
    N=len(x)
    for xi, yi in zip(x,y):
        i =  np.random.randint(0,7)
        image=ims[i]
        im = OffsetImage(image, zoom=10/ax.figure.dpi)
        im.image.axes = ax

        ab = AnnotationBbox(im, (xi,yi), frameon=False, pad=0.0,)

        ax.add_artist(ab)
    pi=0
    for i in range(len(x)):
        if ( np.sqrt( (x[i]-0.5)**2+(y[i]-0.5)**2 )<0.5 ): pi+=1.0
    print(4*pi/N)

    

```

```python

fig, ax = plt.subplots( figsize=(10,10))
Nriso=1000
x = np.random.random(Nriso)
y = np.random.random(Nriso)
ax.scatter(x,y,marker="")
plot_images(x, y, ims, ax=ax)
ax.add_patch( matplt.patches.Circle((0.5,0.5),radius=2   ,color="lightskyblue") )
ax.add_patch( matplt.patches.Circle((0.5,0.5),radius=0.5 ,color="tab:blue") )
eps=0.0
plt.xlim(0-eps,1+eps)
plt.ylim(0-eps,1+eps)
plt.show()
```
